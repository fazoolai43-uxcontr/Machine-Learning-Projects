```python
import numpy as np
import pandas as pd
```

```python
df_water = pd.read_csv("water_potability.csv")
```

```python
print("Sum of null values.")
df_water.isna().sum()
```

```python
df_clean = df_water.fillna(df_water.mean())
```

```python
df_clean
```

```python
df_clean.info()
```

```python
df_clean[['ph','Sulfate','Trihalomethanes']]
```

```python
df_clean.head
```

```python
object_columns = ['ph','Sulfate','Trihalomethanes']
for col in object_columns:
    df_clean[col] = df_clean[col].astype('float64')
```

```python
df_clean.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
sns.heatmap(df_clean.corr(), annot=True)
```

Features Selection

```python
col_list = list(df_clean.columns)[0:-1]
```

```python
x = df_clean[col_list]
```

```python
y = df_clean['Potability']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2)
```

```python
y.value_counts().sum()
```

```python
print('Training input:', x_train.shape)
print('Training output:', y_train.shape)
print('Testing Input:',x_test.shape)
print('Testing output:',y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

print('Accuracy',round(accuracy_score(y_test,pred)*100))
```

```python
plt.figure(figsize=(5,3))
sns.heatmap(confusion_matrix(y_test,pred),annot=True, fmt="")
```
